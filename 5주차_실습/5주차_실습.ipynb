#### 전처리

```python
## 모든 피처별 각한개씩의 결측치가 존재헀다.
df_train.loc[df_train["Age"].isnull()]
```

```python
## 존재하는 결측치는 모든 피처들의 정보가 비어있는 없는 행이므로 이 행은 삭제하는 것으로 결정.
df_train.dropna(ignore_index= True, inplace=True)
```

```python
## 이진 변수 인코딩
df_train["Gender"] = df_train["Gender"].apply(lambda x: 1 if x == "Male" else 0)
```

```python
## 범주형 변수 원핫 인코딩 
encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(df_train[["Subscription Type", "Contract Length"]])

## 데이터 프레임 변환
encoded_data = pd.DataFrame(encoded, columns= encoder.get_feature_names_out())
encoded_data
```

```python
## 데이터 결합
df = pd.concat([df_train, encoded_data], axis = 1)
```

```python
## 기존칼럼 삭제
df.drop(labels= ["Subscription Type","Contract Length"], axis= 1, inplace= True)
```

```python
df
```

- 테스트 데이터 전처리

```python
df_test.isnull().sum()
```

- 테스트데이터에는 결측치가 존재하지 않는다.

```python
## 이진 변수 인코딩 및 원핫인코딩
df_test["Gender"] = df_test["Gender"].apply(lambda x: 1 if x == "Male" else 0)

test_encoded = encoder.transform(df_test[["Subscription Type", "Contract Length"]])
test_en_dat = pd.DataFrame(test_encoded, columns= encoder.get_feature_names_out())
```

```python
## 데이터 결합
test = pd.concat([df_test, test_en_dat], axis= 1)
```

```python
## 기존 칼럼 삭제
test.drop(labels= ["Subscription Type", "Contract Length"], axis= 1, inplace=True)
test
```

#### 모델링

```python
X = df.drop(labels= ["Churn", "CustomerID"], axis= 1)
y = df["Churn"]
```

```python
X_test = test.drop(labels= ["Churn","CustomerID"], axis= 1)
y_test = test["Churn"]
```

```python
## 훈련데이터와 벨리데이션 셋으로 분리
X_train, X_cv, y_train, y_cv = train_test_split(X,y, test_size=0.2, random_state=1234)
```

```python
## 랜덤 포레스트 알고리즘을 통한 훈련
model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
```

```python
print("Train set :", accuracy_score(y_train, model_rf.predict(X_train)))
print("Validation set :",accuracy_score(y_cv, model_rf.predict(X_cv)))
```

```python
print("test score", accuracy_score(y_test, model_rf.predict(X_test)))
```

- 검증 셋과 훈련 셋간의 정확도 차이가 크지 않은 것을 봤을 때 과적합의 문제는 아닌듯 하다

```python
test_pred = model_rf.predict(X_test)
print((test_pred == 0).sum())
test_pred
```

- 거의 모든 데이터를 1로 에측하는 문제

```python
## 로지스틱 회귀 모델
model_lg = LogisticRegression()
model_lg.fit(X_train, y_train)

pred_train = model_lg.predict(X_train)
pred_test = model_lg.predict(X_cv)

print(accuracy_score(y_train, pred_train))
print(accuracy_score(y_cv, pred_test))
```
